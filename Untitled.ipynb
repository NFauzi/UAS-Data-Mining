```python
# Import necessary libraries
import pandas as pd
import math
```

### Exploratory Data Analysis (EDA)
### Melakukan analisis awal terhadap dataset untuk memahami struktur dan statistik descriptif

```python
# Load dataset
dataset = pd.read_csv ('food_full.csv')
```

```python
dataset
```

### Checking for Missing Values
#### Using `isnull().sum()` to Find Missing Values

```python
dataset.isnull().sum()
```

### Handling Missing Values
#### Using `fillna()` to Replace Missing Values

```python
dataset['Sweetener'].fillna(math.floor(dataset['Sweetener'].mean()), inplace= True)
dataset['Biscuits'].fillna(math.floor(dataset['Biscuits'].mean()), inplace= True)
dataset['Jam'].fillna(math.floor(dataset['Jam'].mean()), inplace= True)
```
